### Percobaan 1 : SVM & TF-IDF

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline

df_train = pd.read_excel("/content/dataset_label_manual.xlsx")
df_test = pd.read_csv("/content/data_test (2).csv")

X_train = df_train["komentar"]
y_train = df_train["label_manual"]


model = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('svm', LinearSVC())
])

model.fit(X_train, y_train)

labeling = model.predict(df_test["komentar"])

df_test["label"] = labeling

df_test.to_csv("hasil_labeling_svm_tfidf.csv", index=False)
```
